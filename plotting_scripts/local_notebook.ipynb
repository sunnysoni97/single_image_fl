```python
def plot_accuracy(lines_list, lines_labels, n_rounds, algo_name, dataset_name):
    
    x = [x+1 for x in range(n_rounds)]
    for i in range(len(lines_list)):
        plt.plot(x, lines_list[i], label=lines_labels[i])
    
    plt.xlabel("Round")
    plt.ylabel("Accuracy/100")
    plt.title(f"{algo_name} - {dataset_name}")
    plt.legend()
    plt.show()
```
